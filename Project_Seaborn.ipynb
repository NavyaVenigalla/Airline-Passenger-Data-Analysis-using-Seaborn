```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Flights dataset
flights=sns.load_dataset("flights")

#Set theme
sns.set_theme(style="whitegrid")

#Creating figure with sublots
fig,axes = plt.subplots(2, 2, figsize=(14,10))

#1-Yearly Trend of Airline Passengers (Line Plot)
sns.lineplot(data=flights, x="year", y="passengers", estimator="sum", errorbar=None, marker="o", ax=axes[0,0])
axes[0,0].set_title("Total Passengers Per Year")

#2-Monthly Passenger Distribution Over Years (Box Plot)
sns.boxplot(data=flights, x="month", y="passengers", hue="month", palette="coolwarm", legend=False, ax=axes[0,1])
axes[0,1].set_title("Monthly Passenger Distribution")


#3-Average monthly passenger Count (Bar Plot)
sns.barplot(data=flights, x="month", y="passengers", hue="month", estimator="mean", palette="Blues", ax=axes[1,0])
axes[1,0].set_title("Average Monthly Passengers")

#4-Monthly Passenger Count Over Years(Heatmap)
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
sns.heatmap(flights_pivot,annot=True, fmt="d", cmap="coolwarm", linewidth=0.5, ax=axes[1,1])
axes[1,1].set_title("Heatmap of Monthly Passengers")

#Adjust layout and display
plt.tight_layout()
plt.show()
```
